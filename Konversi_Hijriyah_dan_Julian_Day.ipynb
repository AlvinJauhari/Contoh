```python
print('    KONVERSI HIJRIAH KE JULIAN DAY')
print('======================================')

#Memasukkan kalender hijriah
D = int(input('Masukkan Tanggal: '))
M = input('Masukkan Bulan: ')
Y = int(input('Masukkan Tahun: '))

#Mengubah nama bulan menjadi angka
if M == 'Muharram':
  M1 = 1
  
elif M == 'Safar':
  M1 = 2

elif M == 'Rabiul Awwal':
  M1 = 3
  
elif M == 'Rabiul Akhir':
  M1 = 4

elif M == 'Jumadil Awwal':
  M1 = 5

elif M == 'Jumadil AKhir':
  M1 = 6

elif M == 'Rajab':
  M1 = 7

elif M == 'Syaban:':
  M1 = 8

elif M == 'Ramadhan':
  M1 = 9

elif M == 'Syawal':
  M1 = 10

elif M == 'Dzulqadah':
  M1 = 11
  
elif M == 'Dzulhijjah':
  M1 = 12

Y1 = Y-1
Y2 = int(Y1/30)
Y3 = Y1-(Y2*30)

#Menghitung jumlah tahun kabisat
if Y3 < 2:
  k = 0

elif Y3 < 5:
  k = 1

elif Y3 < 7:
  k = 2
  
elif Y3 < 10:
  k = 3

elif Y3 < 13:
  k = 4

elif Y3 < 16:
  k = 5

elif Y3 < 19:
  k = 6

elif Y3 < 21:
  k = 7

elif Y3 < 24:
  k = 8

elif Y3 < 26:
  k = 9

elif Y3 < 29:
  k = 10

else:
  k = 11

#Menghitung jumlah hari dalam tahun
D1 = Y2*10631
D2 = Y3*354+k

#Mencari jumlah hari pada bulan genap dan ganjil
M2 = M1-1
if M2 %2 == 0:
  D3 = ((M2/2)*30)+((M2/2)*29)

else:
  D3 = (((M2+1)/2)*30)+(((M2-1)/2)*29)

#Jumlah hari
D4 = D+D1+D2+D3

#Nilai JD
JD = 1948438.5+D4

#Output
print(k)
print('\nJadi JD',D,M,Y,'H =',JD)
```

```python
print('    KONVERSI JULIAN DAY KE HIJRIYAH')
print('======================================')

#Masukkan nilai JD
JD = float(input('Masukkan nilai Julian Day: '))

JD1 = JD-1948438.5
A = int(JD1/10631)
B = A*10631
C = JD1-B

#Menghitung t1
Y1 = A*30
#Menghitung t2
Y2 = int(C/354)

#Menghitung jumlah tahun kabisat
if Y2 < 2:
  k = 0

elif Y2 < 5:
  k = 1

elif Y2 < 7:
  k = 2
  
elif Y2 < 10:
  k = 3

elif Y2 < 13:
  k = 4

elif Y2 < 16:
  k = 5

elif Y2 < 19:
  k = 6

elif Y2 < 21:
  k = 7

elif Y2 < 24:
  k = 8

elif Y2 < 26:
  k = 9

elif Y2 < 29:
  k = 10

else:
  k = 11

#Menghitung Hari
D = int(C-(Y2*354+k))
#Menghitung Tahun
Y = Y1+Y2+1

#Mencari Bulan
if D < 31:
  M = 'Muharram'
  D1 = D
  
elif D < 60:
  M = 'Safar'
  D1 = 59 - D
  
elif D < 90:
  M = 'Rabiul Awwal'
  D1 = 89 - D
  
elif D < 119:
  M = 'Rabiul Akhir'
  D1 = 118 - D
  
elif D < 149:
  M = 'Jumadil Awwal'
  D1 = 148 - D
  
elif D < 178:
  M = 'Jumadil Akhir'
  D1 = 177 - D
  
elif D < 208:
  M = 'Rajab'
  D1 = 207 - D
  
elif D < 237:
  M = 'Syaban'
  D1 = 236 - D
  
elif D < 267:
  M = 'Ramadhan'
  D1 = 266 - D
  
elif D < 296:
  M = 'Syawal'
  D1 = 295 - D

elif D < 326:
  M = 'Dzulqaidah'
  D1 = 325 - D
  
elif D < 356:
  M = 'Dzulhijjah'
  D1 = 355 - D

#Output
print('\nJadi JD',JD,'=',D1,M,Y,'H')
```
